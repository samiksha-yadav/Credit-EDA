```python
#import necesssary librarries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', 73)
pd.set_option('display.max_rows', 100)
```

```python
# read csv file and storing it into (apli)--> variable 
apli = pd.read_csv('application_data.csv')
```

```python
# overlook of data by head function
apli.head()
```

```python
apli['CODE_GENDER'].unique()
```

```python
# looking for null values if have and to_string function show the whole null values 
print((100*apli.isnull().mean()).sort_values().to_string())
```

```python
#these are the columns before we drop the null values
apli.shape
```

```python
#taking the thresh for dropping of null columns having values grater than 40%
perc = 40.0
min_count = int(((100-perc)/100)*apli.shape[0]+1)
```

```python
# now we drop the columns and save the dataset into new var i.e mod_apli----> modified
mod_apli = apli.dropna(axis = 1,thresh = min_count)
```

```python
# after dropping the column from 122 it now only upto 73
mod_apli.shape
```

```python
# again watching the null values to modify
print((100*mod_apli.isnull().mean()).sort_values().to_string())
```

```python
# we are imputing the columns we needed other we are droping without imputing...
```

```python
mod_apli['OCCUPATION_TYPE'].unique
mod_apli['OCCUPATION_TYPE'].mode()[0]
mod_apli['OCCUPATION_TYPE'] = mod_apli['OCCUPATION_TYPE'].fillna('Laborers')
```

```python
mod_apli['NAME_TYPE_SUITE'] = mod_apli['NAME_TYPE_SUITE'].fillna('Unaccompanied')
```
